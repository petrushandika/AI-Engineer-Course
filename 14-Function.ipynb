```python
def another(b):
    result = b + 10
    return result
```

```python
another(20)
```

```python
# What different from print vs return?
# print = does not affect the calculation of the output
# return = does not visualize the output, it specifies what a certain function is supposed to give back
```

```python
# Using a Function in Another Function
```

```python
def age(hours):
    return hours * 25

def bonus(hours):
    return age(hours) + 50
```

```python
age(8), bonus(8)
```

```python
# Combining Conditional Statements and Functions
```

```python
def add(m):
    if m >= 100:
        m = m + 10
        return m
    else:
        return "Save More!"
```

```python
add(110)
```

```python
add(50)
```

```python
# Creating Functions Containing a Few Arguments
```

```python
def subtract(a, b, c):
    result = a - b * c
    print("a = ", a)
    print("b = ", b)
    print("c = ", c)
    return result
```

```python
subtract(10, 3, 2)
```
