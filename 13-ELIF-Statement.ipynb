```python
def compare(y):
    if y > 5:
        return "Greater"
    elif y < 0:
        return "Negative"
    elif y < 5:
        return "Less"
    else:
        return "Equal"
```

```python
print(compare(10))
```

```python
print(compare(3))
```

```python
print(compare(5))
```

```python
print(compare(-1))
```
